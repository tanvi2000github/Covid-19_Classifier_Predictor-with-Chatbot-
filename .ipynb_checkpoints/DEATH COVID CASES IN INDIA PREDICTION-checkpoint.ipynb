## Accuracy Of Predictions In Test Set

```python
x=india["Date"]#.values.astype(int).reshape(-1,1)
y=india["Death"]#.values.reshape(-1,1)
x=x.values.astype('datetime64[D]').reshape(-1,1).astype(int)
x_train=x[0:-30]
y_train=y[0:-30]
y_test=y[-30:]
x_test=x[-30:]

from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x_train)
poly.fit(x_poly,y_train)
y_pred=poly.fit_transform(x_test)
from sklearn.linear_model import LinearRegression
lr1=LinearRegression(fit_intercept=False)
lr1.fit(x_poly,y_train)
y_pred=lr1.predict(y_pred).astype(int)
y_pred
```

```python
import sklearn.metrics as sm
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2)) 
print("R2 score =", sm.r2_score(y_test, y_pred))
```

## Predicting Future Covid Cases In India

```python
x=india["Date"]#.values.astype(int).reshape(-1,1)
y=india["Death"]#.values.reshape(-1,1)
x_test=x.values.astype('datetime64[D]').reshape(-1,1)
last_date=x_test[-1]
new_dates = []
for i in range(1,8):
    next_date=last_date+i
    new_dates.append(next_date)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x_test)
poly.fit(x_poly,y)
y_pred=poly.fit_transform(new_dates)
lr1=LinearRegression(fit_intercept=False)
lr1.fit(x_poly,y)
final=lr1.predict(y_pred)
final=final.astype(int)
```

```python
output=pd.DataFrame(zip(new_dates,final) , columns= ["Future Dates","Future Death Cases"])
output['Future Dates']=[x[0] for x in output['Future Dates']]
output
```

```python
fig = go.Figure()
fig.add_traces( go.Scatter(x=x, y=y, name=" Present Death Cases",marker_color='orange',mode='lines+markers'))
fig.add_traces(go.Scatter(x=output['Future Dates'],y=output['Future Death Cases'] ,name=" Future Death Cases" ))
fig.update_layout(title_text = 'Actual and Predicted Death Cases')
fig.update_layout(template='plotly_dark',width=1000, height=600,xaxis_title='Dates',yaxis_title='Death Cases')
fig.show()
```
